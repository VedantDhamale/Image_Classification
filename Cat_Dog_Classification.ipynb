```python
import pandas as pd
import numpy as np
import scipy
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip= True)

training_set = train_datagen.flow_from_directory('Cat_and_Dog/train',
                                                 target_size = (64,64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip= True)
test_set = test_datagen.flow_from_directory('Cat_and_Dog/test',
                                                 target_size = (64,64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```python
# Pooling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units= 128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units= 1,activation='sigmoid'))
```

```python
cnn.summary()
```

```python
# Training the CNN
cnn.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 200)
```

```python

test_image = tf.keras.utils.load_img('Cat_and_Dog/test/dogs/dog_555.jpg',target_size= (64,64))
```

```python
test_image = tf.keras.utils.img_to_array(test_image)
```
